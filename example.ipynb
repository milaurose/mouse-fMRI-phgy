```python
path_to_nifti = "../fmri_piloting_project/4_derivatives/phgy_fmri_project/rabies_commonspace_template_mask/resampled_template.nii.gz"
```

# Load using nibabel

```python
import nibabel as nb
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the file
nifti_obj = nb.load(path_to_nifti)
nifti_obj
```

```python
#convert to array
nifti_array = np.asarray(nifti_obj.dataobj)
nifti_array.shape
```

```python
#plot image arrays using imshow
plt.imshow(nifti_array[:,80,:])
```

# Load using sitk

```python
import SimpleITK as sitk
```

```python
nifti_obj_v2 = sitk.ReadImage(path_to_nifti)
nifti_obj_v2
```

```python
#convert to array
nifti_array_v2 = sitk.GetArrayFromImage(nifti_obj_v2)
nifti_array_v2.shape
```

```python
#plot image arrays using imshow
plt.imshow(nifti_array_v2[:,80, :])
```

```python
#plot image arrays using imshow
plt.imshow(nifti_array_v2[:,80, :], origin='lower')
```
